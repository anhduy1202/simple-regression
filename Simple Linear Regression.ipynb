# Simple Regression Equation:

##  

```python
df[['sales','total_spend']]
```

```python
y_mean = df['sales'].mean()
x_mean = df['total_spend'].mean()
df['y-y_mean'] = df['sales'] - y_mean
df['x-x_mean'] = df['total_spend'] - x_mean
df['(x-x_mean)(y-y_mean)'] = df['x-x_mean'] * df['y-y_mean']
df['(x-x_mean)^2'] = df['x-x_mean'] ** 2
b1 = round(df['(x-x_mean)(y-y_mean)'].sum() / df['(x-x_mean)^2'].sum(),2)
b0 = round(y_mean - b1 * x_mean,2)
```

```python
x = np.linspace(0,500,100)
```

```python
y = b0 + b1 * x
```

```python
plt.figure(figsize=(10,8))
plt.title("Simple Regression Sales vs Total Spend",fontsize=16)
sns.scatterplot(x='total_spend',y='sales',data=df);
plt.plot(x,y,color="red");
```
